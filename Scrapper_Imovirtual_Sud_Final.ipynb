```python
def scrape_content(url):
    response_is_successful = False
    attempts = 0
    while response_is_successful == False:
        global proxyURL
        headers = {
            'authority': 'www.imovirtual.com',
            'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
            'accept-language': 'en-US,en;q=0.9',
            # Requests sorts cookies= alphabetically
            # 'cookie': 'laquesisff=gre-12226#rer-11#rer-181#rer-182; lqstatus=1656412694; OptanonAlertBoxClosed=2022-06-28T10:18:20.001Z; OTAdditionalConsentString=1~89.2008.2072.2322.2465.2501.2999.3028.3225.3226.3231.3232.3234.3235.3236.3237.3238.3240.3241.3244.3245.3250.3251.3253.3257.3260.3268.3270.3272.3281.3288.3290.3293.3295.3296.3300.3305.3307.3308.3314; optimizelyEndUserId=oeu1656411500686r0.4104159372927598; __gads=ID=5e8fb439dbb6d926:T=1656411503:S=ALNI_Mb-nPxeRtcLViiUjwMSkuYQvC9Wmg; _ga=GA1.2.1880454660.1656411500; _gcl_au=1.1.786952406.1656411502; laquesis=; laquesissu=315@my_account|0; __diug=true; __gfp_64b=KeQbvjzFi3zUL3u_sQy.ZSLXd3o5d0ODhVHPE1lJeir.E7|1656411505; newrelic_cdn_name=CF; PHPSESSID=0718ones8v4hj9959q4i399l5h; mobile_default=desktop; ninja_user_status=unlogged; observed5_id_clipboard=62de2af4d02f4; observed5_sec_clipboard=duMW4Q%2Ff48rIdVYWJlkQevt4kr7OPSMCQSn8JN4yOtM%3D; OptanonConsent=isGpcEnabled=0&datestamp=Mon+Jul+25+2022+10%3A32%3A37+GMT%2B0500+(Pakistan+Standard+Time)&version=6.21.0&isIABGlobal=false&consentId=69e4df12-01c9-424e-862d-a1a34a8e4a80&interactionCount=1&landingPath=NotLandingPage&groups=C0001%3A1%2CC0002%3A1%2CC0003%3A1%2CC0004%3A1%2Cgad%3A1%2CSTACK42%3A1&hosts=H234%3A1%2Capp%3A1%2CH39%3A1%2CH514%3A1%2CH139%3A1%2CH1%3A1%2CH8%3A1%2CH120%3A1%2CH9%3A1%2CH478%3A1%2CH232%3A1%2CH13%3A1%2CH278%3A1%2CH214%3A1%2CH332%3A1%2CH167%3A1%2CH458%3A1%2CH169%3A1%2CH204%3A1%2CH125%3A1%2CH171%3A1%2CH33%3A1%2CH173%3A1%2CH38%3A1%2CH175%3A1%2CH41%3A1%2CH44%3A1%2CH178%3A1%2CH179%3A1%2CH46%3A1%2CH479%3A1%2CH206%3A1%2CH469%3A1%2CH238%3A1%2CH133%3A1%2CH471%3A1%2CH187%3A1%2CH221%3A1%2CH70%3A1%2CH188%3A1%2CH189%3A1%2CH280%3A1%2CH73%3A1%2CH191%3A1%2CH79%3A1%2CH192%3A1%2CH209%3A1%2CH334%3A1%2CH513%3A1%2CH194%3A1%2CH227%3A1%2CH196%3A1%2CH95%3A1%2CH97%3A1%2CH144%3A1%2CH197%3A1%2CH473%3A1%2CH281%3A1%2CH112%3A1%2CH462%3A1%2CH542%3A1%2Clim%3A1&genVendors=V8%3A1%2C&geolocation=PK%3BPB&AwaitingReconsent=false; eupubconsent-v2=CPbTFYRPbTFYRAcABBENCZCsAP_AAH_AAAYgI3Nf_X__b3_j-_5_f_t0eY1P9_7__-0zjhfdl-8N3f_X_L8X42M7vF36pq4KuR4Eu3LBIQdlHOHcTUmw6okVrzPsbk2cr7NKJ7PEmnMbO2dYGH9_n13TuZKY7_____7z_v-v_v____f_7-3f3__p_3_--_e_V_99zfn9_____9nP___9v-_9__________3wRrAJMNW4gC7EscCTaMIoUQIwrCQ6gUAFFAMLRBYQOrgp2VwE-oIWACAVARgRAgxBRgwCAAQCAJCIgJADwQCIAiAQAAgAVAIQAEbAILACwMAgAFANCxAigCECQgyICI5TAgIkSCgnsrEEoO9DTCEOssAKBR_RUICJQAgWBkJCwchwBICXCyQLMUL5ACMEKAUAAAAA.f_gAD_gAAAAA; __gpi=UID=0000096eb61d30b8:T=1656411503:RT=1658727161:S=ALNI_MbDZmuKxeozJgQ7gQzy1lIDUFEc8g; ldTd=true; _gid=GA1.2.1593321118.1658727163; _gat_clientNinja=1; lqstatus=1658728363||||; cto_bundle=Jdt-BV9PQnc4ZSUyRkhxMnVibGN5UFhwTWVLbDFuY1c5JTJGQzVKcmxicDJRYiUyQlFyTHVFU0RLNTF0YjJGaGVBREpWTG4lMkJyYnpZRTFUVUYwcVJTTzhZS05UJTJCbjh2UmJtUGZjb0Vwb1RYWSUyRmpDckVmZ0xDUEZvd1ZYSm1ubVRhMWMxU3NzTENtWm03ZSUyQmg2cU1lJTJGczV3NW5ka1NxemV3JTNEJTNE; onap=f10a22fede8x59511c0d-13-18233d7df08x38d3066b-4-1658728969',
            'sec-ch-ua': '".Not/A)Brand";v="99", "Google Chrome";v="103", "Chromium";v="103"',
            'sec-ch-ua-mobile': '?0',
            'sec-ch-ua-platform': '"Windows"',
            'sec-fetch-dest': 'document',
            'sec-fetch-mode': 'navigate',
            'sec-fetch-site': 'same-origin',
            'sec-fetch-user': '?1',
            'upgrade-insecure-requests': '1',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36',
        }
        headers["referer"] = url

        opener = urllib.request.build_opener(
        urllib.request.ProxyHandler(
            {'http': str(proxyURL),
            'https': str(proxyURL)}))
        opener.addheaders = headers.items()

        
        try:
            response = opener.open(url, timeout=10)
            response_content = response.read()
            if response.code == 200 and response_content != "" and response_content != None:
                print(colored('Page Loaded : OK' , "green"))
                response_is_successful = True
                return True, response_content # SUCCESS, Html text
            else:
                raise Exception(f'response code was {response.code}')
            
        except Exception as e:
            print(colored(f'Page Loaded : NO' , "red"))
            attempts += 1
            print(f"Attempt {attempts}/10")
            print("Exception:", e)
            print("Waiting 3 seconds...")
            time.sleep(3)
            if attempts == 10:
                print('Moving to the next page ! (We cannot try more than 10 times)')
                return False,"" # NO SUCCESS, EMPTY text

def get_one_details_page(p_ind, total_porp, link):
    print(colored(f"Retrieving {p_ind+1}/{total_porp}......." , 'yellow'))
    was_successful, response_content = scrape_content(link)
    if was_successful:
        soup = BeautifulSoup(response_content, "lxml")
        property = {}
        try:property["Title"] = soup.find("h1", {"data-cy":"adPageAdTitle"}).text
        except: property["Title"] = ""
        try:property["Price"] = soup.find("strong", {"data-cy":"adPageHeaderPrice"}).text
        except: property["Price"] = ""
        try:property["Image"] = "\""+soup.find("picture").img["src"]+"\""
        except: property["Image"] = ""
        try:property["Link"] = link
        except: property["Link"] = ""
        temp_tab = [tab.text.replace(".css-9c7r58{font-size:10px;}.css-1q4vxyr{margin-left:8px;}", "") 
                        for tab in soup.findAll("a", {"class", "css-1in5nid e1je57sb4"})]

        ######################################################################## New ########################################################################
        try:
            # Getting Article
            main_article = soup.find('main').find('div', {'class':'css-9j501y'})
            # Getting navigation bar Div
            tab_soup = main_article.find_all('div')[0]
            # Getting all a tags
            a_tags_content = []
            for a_tag in tab_soup.find_all('a'):
                a_tags_content.append(a_tag.text)

            # Checking if we have 5 fields therefore we can conclude that this NavBar has street info in it.
            has_street_info = False
            if len(a_tags_content) == 5:
                has_street_info = True
            
            # Saving
            try:
                typeAppart, dealAppart = a_tags_content[0].split('para')
                property["Type"] = typeAppart.strip().capitalize()
                property["Deal"] = dealAppart.strip().capitalize()
            except:
                property["Type"] = ""
                property["Deal"] = ""
            property["Distro"] = a_tags_content[1]
            property["Cidade"] = a_tags_content[2]
            property["Freguesia"] = a_tags_content[3]
            if has_street_info:
                property["Rua"] = a_tags_content[4]

            # Getting Header Section (where title and address are located)
            try:
                # Getting Header
                header_soup = main_article.find_all('div' ,recursive=False)[1].find('header')
                # Getting Adress Div
                address_soup = header_soup.find('a')
                property["Address"] = address_soup.text
            except:
                property["Address"] = ""
        except:
            print('Error in the NavBar Data Scraping')
        ######################################################################## New ########################################################################
        try:property["Description"] = soup.find("div", {"data-cy":"adPageAdDescription"}).text.replace("\n", " ")
        except: property["Description"] = ""
        p_properties_section = soup.find("div", {"class":"css-shnn4j e17tijfc0"})
        try:
            for p_prop in p_properties_section.findAll("div", {"role":"region"}):
                temp = [i.text for i in p_prop.findAll("div")]
                property[temp[0]] = temp[1]
        except: pass
        try:property["Features"] = [i.text for i in soup.find("ul", {"data-cy":"ad.ad-features.uncategorized-list"}).findAll("li")]
        except: property["Features"] = []
        
        
        return property
    else:
        return None

def get_properties_by_list(p_links, total_number_of_old_df, total_number_of_articles):
    total_porp = total_number_of_articles
    all_properties = []
    for (p_ind, link) in enumerate(p_links):
        property = get_one_details_page(p_ind + total_number_of_old_df, total_porp, link)
        if property != None:
            all_properties.append(property)
            #time.sleep(random.uniform(1.6, 2.8))
            #time.sleep(random.uniform(0.8, 1.4))
            #time.sleep(random.uniform(0.4, 0.7))
            time.sleep(random.uniform(0.2, 0.35))
            
            intervalos = []#[5000,10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000]
            if p_ind+1 in intervalos:
                #print('OLA')
                time.sleep(360)
        else:
            print("PP")

    return all_properties

def get_number_of_offers(c_link):
    was_successful = False
    while was_successful == False:
        was_successful, response_content = scrape_content(f"{c_link}?nrAdsPerPage=72&page=1")
        if was_successful:
            soup = BeautifulSoup(response_content, "lxml")
            try:
                n_offers = int(soup.find('div', {'class':'offers-index pull-left text-nowrap'}).find('strong').text.replace(" ", "").strip())
                print(f"Found {n_offers} offers")
                return n_offers
            except:
                was_successful = False


def get_articles_from_listpage(c_link, number_of_pages):
    links = []
    cities = []

    for page in range(1,number_of_pages+1) :
        was_successful = False
        while was_successful == False:
            try:
                was_successful, response_content = scrape_content(f"{c_link}?nrAdsPerPage=72&page={page}")
            except:
                pass
            if was_successful:
                soup = BeautifulSoup(response_content, "lxml")

                for i in soup.findAll("article"):
                    link = address = ''
                    link = i.find("a")["href"]
                    address = i.find("p", {"class" : "text-nowrap"}).text.split(':')[1]
                    resultAddress = re.split(r',(?![^(]*\))', address)
                    city = resultAddress[len(resultAddress)-1].strip()

                    links.append(link)
                    cities.append(city)
                print(colored(f"Found {len(links)} Articles on Page {page}, Searching for More..." , "yellow"))
    return links, cities

def filter_data(dataframe, filter_key, filter_value):
    df = dataframe[dataframe[filter_key]==filter_value]
    return df


def scrape_category(all_catagories, cat_id, filter_cities, cities_to_include):
    success = False
    while success == False:
        # Init file name
        cat_link = all_catagories[cat_id]
        current_time = time.time()
        formatted_time = time.strftime("_Created on (%d %m %Y %H-%M-%S)", time.localtime(current_time))
        cat_name = "_".join(list(reversed(cat_link.split("/")[-3:-1]))) + formatted_time + ".CSV"
        print(colored(f"We will be scraping for this category : {cat_link}" , "blue"))

        # Checking ads number
        number_of_offers_found = get_number_of_offers(cat_link)
        number_of_pages = int(number_of_offers_found/72) + (number_of_offers_found/72 > int(number_of_offers_found/72))
        print(f'This Category has {number_of_pages} pages (72 articles per page)')
        p_links, p_cities = get_articles_from_listpage(cat_link, number_of_pages)
        print('Our search results contains these cities :')
        for city, count in Counter(p_cities).items():
            print(f'{city} has {count} Offers')

        articles = []
        for i in range(len(p_links)):
            articles.append({'link': p_links[i], 'city' : p_cities[i]})

        print(colored(f"Got {len(p_links)} Properties from this Catagory." , "blue"))

        full_links_list = []
        # Filter 
        for article in articles:
            if filter_cities:
                if article.get('city') in cities_to_include:
                    full_links_list.append(article.get('link'))
            else:
                full_links_list.append(article.get('link'))
        
        # creating a new list that contains sublists of 100 entities each from full_links_list
        list_of_sublists_100 = [full_links_list[i:i+100] for i in range(0, len(full_links_list), 100)]

        if len(list_of_sublists_100) > 0:
            success = True
        
        # Saving chunk by chunk (100 Ads per chunk)
        print(colored("Started Scraping...!"  ,'red'))

        print(f'{len(list_of_sublists_100)} chunks !')

        df = None
        for i in range(len(list_of_sublists_100)):
            if df is None:
                len_df = 0
            else:
                len_df = len(df)
            # Scraping
            new_coming_data_df = pd.DataFrame(get_properties_by_list(list_of_sublists_100[i], len_df, len(full_links_list)))

            # Saving
            if df is None:
                df = new_coming_data_df
            else:
                df = pd.concat([df, new_coming_data_df], ignore_index=True)

            try: # In the case of the csv file was open during the script then we pass
                df.to_csv(cat_name, index=False, encoding="utf-8-sig")
                print(colored(f"Chunk Completed and saved to {cat_name}\n"  ,'blue'))
            except:
                print(colored(f"Chunk Completed but not saved to {cat_name}\n"  ,'red'))

        print(colored("Catagory Completed!\n"  ,'blue'))
        #all_df.pop(df1)
    return df
```
